```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/portland.csv')
```

```python
#<ipython-input-1-4becc0fdf2b1>:4: DtypeWarning: Columns (2,11,12) have mixed types. Specify dtype option on import or set low_memory=False.
  #df = pd.read_csv('/content/portland.csv')

# columns 2, 11, and 12 have mixed types - change them all to be object

df[['city', 'owner', 'legal_description']] = df[['city', 'owner', 'legal_description']].astype('object')
```

```python
# change sale date to datetime type

df['sale_date'] = pd.to_datetime(df['sale_date'])
df['year_built'] = pd.to_datetime(df['year_built'])
```

```python
df.dtypes
```

```python
# drop rows with at least one null value
df.dropna(inplace = True)
df
```

```python
# create a new dataframe that only shows information for primary residences
primary_residence = df[df['primary'] == 1]
primary_residence.head()
```

```python
# years with the most sales of primary residences - count
sales_by_year = primary_residence.groupby(primary_residence['sale_date'].dt.year)['property_id'].count().sort_values(ascending = False)
sales_by_year.head(10)
```

```python
sale_amount_by_year = primary_residence.groupby(df['sale_date'].dt.year)['sale_price'].sum().sort_values(ascending = False)
sale_amount_by_year.head()
```
