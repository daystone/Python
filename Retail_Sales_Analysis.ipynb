**Objective**

Create a Python script that analyzes a retail sales dataset, focusing on data manipulation with pandas and visualization of sales trends

Tasks

1. Read the Data in a DataFrame
2. Data cleaning
3. Analysis
  - calculate total sales by product
  - identify best selling product
  - compute average daily sales
4. Visualization
  - plot sales trends over time
  - display sales per product in a bar chart
5. Class and function
  - Define a Python class **RetailSalesAnalyzer** with methods for each of the above tasks (loading data, caculating statistics, and creating visuals
6. Script execution
  - Write a script that creates an instance of RetailSalesAnalyzer, calls its methods, and displays the results




```python
#Define a Python class RetailSalesAnalyzer with methods for each of the above tasks

# Import libraries

import pandas as pd
import matplotlib.pyplot as plt

# Create class

class RetailSalesAnalyzer:
  def __init__(self):
    self.data = pd.read_csv('/content/retail_sales.csv')
    self.data['Date'] = pd.to_datetime(self.data['Date'])

# Cleaning Data
  def clean_data(self):
    self.data.dropna(inplace = True)

# Analyze Data
  def total_sales_by_product(self):
    return self.data.groupby('Product')['Sales'].sum()

  def best_seller(self):
    return self.total_sales_by_product().sort_values(ascending = False).index[0]

  def avg_daily_sales(self):
    return self.data.groupby('Date')['Sales'].mean()

# Visualize the Data
  def sales_over_time(self):
    self.data.groupby('Date')['Sales'].sum().plot(kind = 'line')
    plt.title('Sales Over Time')
    plt.xlabel = 'Date'
    plt.ylabel = 'Total Sales'
    plt.show()

  def plot_sales_by_product(self):
    self.total_sales_by_product().plot(kind = "bar")
    plt.title = 'Sales by Product'
    plt.xlabel = 'Product'
    plt.ylabel = 'Total Sales'
    plt.show()

```

```python
analyzer = RetailSalesAnalyzer()
print('Total Sales per Product: \n', analyzer.total_sales_by_product())
print('Best Seller: ', analyzer.best_seller())
print('Average Daily Sales: ', analyzer.avg_daily_sales())

analyzer.sales_over_time()
analyzer.plot_sales_by_product()
```
